# INTRODUCTION OF WATER BODIES IN DELHI.


Delhi, the capital of India, is home to several important water bodies that play a crucial role in the city's ecology and water supply. The major water bodies in Delhi include rivers, lakes, and wetlands, each contributing to the region's environmental health, biodiversity, and water management systems.

### Key Water Bodies in Delhi:
1. **Yamuna River**: The Yamuna is the most significant water body in Delhi, flowing through the city and providing a major source of water. However, the river faces severe pollution challenges due to industrial discharge, untreated sewage, and agricultural runoff, making its water quality a critical concern for both the environment and public health.

2. **Lakes and Reservoirs**: Delhi has several natural and man-made lakes, such as the **Sanjay Lake**, **Hauz Khas Lake**, and **Bhalswa Lake**. These water bodies are important for groundwater recharge, recreational activities, and maintaining local ecosystems. However, many of these lakes suffer from pollution, encroachment, and neglect.

3. **Wetlands**: Wetlands like the **Najafgarh Jheel** and **Okhla Bird Sanctuary** play a vital role in maintaining the ecological balance by supporting diverse flora and fauna. These wetlands are also crucial for flood control, water purification, and serving as habitats for migratory birds.

### Challenges:
Despite their importance, Delhi's water bodies are under significant threat due to rapid urbanization, pollution, and over-extraction of water resources. The degradation of these water bodies has led to declining water quality, loss of biodiversity, and increased vulnerability to water scarcity.

### Importance of Conservation:
The conservation and restoration of Delhi's water bodies are essential for sustaining the city's water supply, improving environmental quality, and preserving biodiversity. Efforts to clean and rejuvenate these water bodies, such as the Yamuna Action Plan, are critical to ensuring a sustainable future for the city.

```python
df.drop("STATE",axis =1,inplace =True)
```

```python
df.head()
```

```python
df["LOCATIONS"]= df["LOCATIONS"].map({'Shahdara Drain':1,'Yamuna at ITO':2,'Yamuna at Nizamuddin':3,'Yamuna at Okhla':4,'Najafgarh Drain':5,'Yamuna at Wazirabad':6})
```

```python
df
```

```python
X = df.drop('WATER QUALITY INDEX', axis =1)
y = df['WATER QUALITY INDEX']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
dtr = DecisionTreeRegressor()
svr = SVR(kernel = "linear")
```

```python
from sklearn.ensemble import VotingRegressor
ensemble_regressor = VotingRegressor(estimators=[('mlr', lr), ("dtr", dtr), ("svr", svr)])
```

```python
ensemble_regressor
```

```python
ensemble_regressor.fit(X_train, y_train)
```

```python
y_pred = ensemble_regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
